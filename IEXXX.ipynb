```python
#SMA

def SMA(close, periods):
    
    sma = np.zeros(periods - 1)
    count = 0

    
    for i in range(periods, len(close) + 1):
        periods_rng = list(range(count, i))
        smaC = close[periods_rng].sum() / periods
        sma = np.append(sma, smaC)
    
        count += 1

    return sma

print(SMA(df["close"], 9))

```

```python
#function for ema calc
#fix function arguments


def ema(close, periods):

 
    sma = close[list(range(0,periods))].sum()/periods
    mult = 2/(periods + 1)

    ema_lst = np.zeros(periods-1)
    ema_lst = np.append(ema_lst,sma)


    for i in range(periods-1, close.size-1):
        ema = ((close[i] - ema_lst[i]) * mult) + ema_lst[i]
        ema_lst = np.append(ema_lst,ema)
  
    return ema_lst



df["ema"] = ema(df["close"],9)

df
```

```python
# RSI


def RSI(close,periods):
    #Gain & Loss
    gain = np.zeros(1)
    loss = np.zeros(1)
    
    for i in range(1, len(close)):
        change = close[i] - close[i - 1]
        
        if change > 0:
            gain = np.append(gain, change)
            loss = np.append(loss, 0)
            
        elif change < 0:
            gain = np.append(gain, 0)
            loss = np.append(loss, abs(change))
            
        else:
            gain = np.append(gain, 0)
            loss = np.append(loss, 0)

            
    #Avg Gain & Avg Loss
    avgG = np.append(np.zeros(periods - 1), np.average(gain[0:periods]))
    avgL = np.append(np.zeros(periods - 1), np.average(loss[0:periods]))
    rsi = np.append(np.zeros(periods - 1), 100 - (100 / (1 + (np.average(gain[0:periods] / np.average(loss[0:periods]))))))

    
    for i in range(periods, len(close)):
        avgG = np.append(avgG, ((avgG[i - 1] * 13) + gain[i]) / 14)
        avgL = np.append(avgL, ((avgL[i - 1] * 13) + loss[i]) / 14)
        rsi = np.append(rsi, 100 - (100 / (1 + (avgG[i] / avgL[i]))))\
        
    return rsi
 
print(RSI(df["close"], 14))
    
    
    
    
```

```python
#Bollinger Bands
from math import sqrt

"""
  * Middle Band = 20-day simple moving average (SMA)
  * Upper Band = 20-day SMA + (20-day standard deviation of price x 2) 
  * Lower Band = 20-day SMA - (20-day standard deviation of price x 2)
"""

def Bollinger(close, periods):
    
    sd = np.zeros(periods - 1)
    
    bolL = np.zeros(periods - 1)
    bolM = SMA(df["close"] , periods)
    bolU = np.zeros(periods - 1)
    
    per_range = list(range(0,periods))
    count = 0
    
    for i in range(periods, len(close)+1):
        sd_per = close[list(range(count, i))]
        sd_per = (sd_per - bolM[i - 1]) ** 2
        sd_per = sqrt(np.average(sd_per))

        sd = np.append(sd, sd_per)

        count += 1
        
        bolL = np.append(bolL, bolM[i-1] - (2 * sd[i-1]))
        bolU = np.append(bolU, bolM[i-1] + (2 * sd[i-1]))
        
    
    
    return (bolL, bolM, bolU)
    
    
Bollinger(df["close"], 20)

        
        
    
        
```
